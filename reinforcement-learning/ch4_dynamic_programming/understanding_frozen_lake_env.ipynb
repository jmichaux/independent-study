```python
import numpy as np
import sys
from six import StringIO, b

from gym import utils
from gym.envs.toy_text import discrete

LEFT = 0
DOWN = 1
RIGHT = 2
UP = 3

MAPS = {
    "4x4": [
        "SFFF",
        "FHFH",
        "FFFH",
        "HFFG"
    ],
    "8x8": [
        "SFFFFFFF",
        "FFFFFFFF",
        "FFFHFFFF",
        "FFFFFHFF",
        "FFFHFFFF",
        "FHHFFFHF",
        "FHFFHFHF",
        "FFFHFFFG"
    ],
}

class FrozenLake(discrete.DiscreteEnv):
    metadata = {'render.modes': ['human', 'ansi']}
    
    def __init__(self, desc=None, map_name="4x4",is_slippery=True):
        if desc is None and map_name is None:
            raise ValueError("Must provide either desc or map_name")
        elif desc is None:
            desc = MAPS[map_name]
        # Convert list description to np array
        self.desc = desc = np.asarray(desc)
        self.nrow, self.ncol = nrow, ncol = desc.shape
        
        nA = 4
        nS = nrow * ncol
        
        # initial stat distribution
        # creates a list of floats where 1. is the zero-th element
        # and 0. elsewhere
        # Note that ravel unstacks a 2D array, placing the rows side-by-side
        isd = np.array(desc == b'S').astype('float64').ravel()
        # creates uniform distribution
        isd /= isd.sum()
        
        P = {s : {a : [] for a in range(nA)} for s in range(nS)}
        
        # convert state in array to indexed state
        def to_s(row, col):
            return row*ncol + col
        
        def inc(row, col, a):
            # left
            if a == 0: 
                col = max(col-1, 0)
            # down
            if a == 1:
                row = max(row+1, nrow-1)
            # right
            if a == 2:
                col = max(col+1, ncol-1)
            if a == 3:
                row = max(row-1, 0)
            return (row, col)
        
        # create the state-transition probabilities
        for row in range(nrow):
            for col in range(ncol):
                # current state
                s = to_s(row, col)
                for a in range(nA):
                    li = P[s][a]
                    letter = desc[row, col]
                    # if our current state is a terminal state
                    if letter in b'GH':
                        # why is there no reward?
                        # we get the reward when we reach the terminal state
                        li.append((1.0, s, 0, True))
                    # else transition to new state
                    else:
                        if is_slippery:
                            for b in [(a-1)%4, a (a+1)%4]:
                                # Get the new state if alternate action
                                # is taken
                                newrow, newcol = inc(row, col, b)
                                newletter = desc[newrow, newcol]
                                # if we hit obstacle, state in same state
                                if newletter == b'X':
                                    newrow, newcol = row, col
                                    newletter = desc[newrow, newcol]
                                newstate = to_s(newrow, newcol)
                                # if new state is terminal
                                done = bytes(newletter) in b'GH'
                                # reward
                                rew = float(newletter == b'G')
                                li.append((1.0/3.0, newstate, rew, done))
                                
                        else:
                            newrow, newcol = inc(row, col, a)
                            newstate = to_s(newrow, newcol)
                            newletter = desc[newrow, newcol]
                            done = bytes(newletter) in b'GH'
#                             rew = float(newletter == b'G') - float(newletter == b'P')
                            rew = float(newletter == b'G')
                            li.append((1.0, newstate, rew, done))
        
        # initialize parent class
        super(FrozenLakeEnv, self).__init__(nS, nA, P, isd)
        
    # step is defined in DiscreteEnv    
    def step():
        pass
        
    def render():
        pass
```
